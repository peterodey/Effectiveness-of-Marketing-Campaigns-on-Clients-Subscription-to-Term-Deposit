## Tittle: Effectiveness of Marketing Campaigns on Client Subscription to Term Deposit

This project centers on understanding the effectiveness of marketing campaigns conducted by a banking institution on client subscription to term deposit. A term deposit refers to money deposit in a banking institution that cannot be withdrawn for a certain "term" or period of time.

## Statement of the Problem

Marketing resources such as budget, time, and manpower are limited. Therefore, understanding the effectiveness of marketing campaigns on client subsription to term deposite will help prioritize leads likely to subscribe to term deposits, allowing banking institutions to allocate resources more efficiently to achieve higher returns on investment.

## Aim and Objectives of the Data Analysis
The aim of this project is to investigate the effectiveness of marketing campaigns on client subscription to Term Deposit.

The specific objectives of this study are:

1. To perform descriptive statistics and exploratory data analysis of the distribution and relationships between variables.

2. To Segment the data into different customer groups respond to marketing campaigns.

3. To Explore temporal trends in marketing campaign success by months and days of the week effects.

4. To Create customer profiles based on job, marital status, education, default, housing, loan, contact, poutcome.

5. To Analyze the influence of contact duration and past campaign outcomes on subscription rates.


## Data Collection and Methodology

The data was obtained from Kaggle, and analysed using python libraries such as Pandas, Matplotlib, seaborn.

## Target Audience

This project will provide insights and recommendations to marketing teams to optimize a marketing campaign strategy and improve subscription rates.

### Descriptive statistics and exploratoration of the distribution and relationships between variables

```python
# Summary of statistics for the numerical columns
df.describe()
```

```python
# Distribution of the target variable 'y'
plt.figure(figsize=(6, 6))

df['y'].value_counts().plot.pie(autopct='%1.1f%%', labels=['No', 'Yes'])
plt.title('Distribution of Subscription to Term Deposit')

# Save picture as pdf and jpg
plt.savefig('artifacts/Distribution_of_Subscription_to_Term_Deposit.pdf', format='pdf')
plt.savefig('images/Distribution_of_Subscription_to_Term_Deposit.jpg')
plt.show()
```

The distribution of Subscription to Term Deposit revealed 5289 (11.7 %) of client subscribed to term deposit while 39922 (88.3 %) client did not subscribe to term deposit

```python
# Distribution of the numerical variables
num_features = ['age', 'balance', 'day', 'duration']

plt.figure(figsize=(15, 10))
for i, feature in enumerate(num_features, 1):
    plt.subplot(3, 3, i)
    sns.histplot(df[feature], bins=20, kde=True, color='skyblue')
    plt.title(f'{feature.capitalize()} vs Subscription')
    plt.xlabel(feature)
    plt.ylabel('Frequency')

plt.suptitle('Distribution of Numerical Features')
plt.tight_layout()

# Save picture as pdf and jpg
plt.savefig('artifact/Distribution_of_Numerical_Features.pdf', format='pdf')
plt.savefig('image/Distribution_of_Numerical_Features.jpg')
plt.show()
```

Distribution of Numerical Features revealed:

1. that aged distribution was highest at 32 years, and lowest at 94 years
2. that account balance had the highest peak at €2000 which was from 3514 clients
3. that day 20 had the highest peak with 4800 contact with clients
4. that last highest contact duration 188 seconds to 29000 clients and lowest at 1 second to 1556 clients

```python
# The relationships between categorical features and the target variable 'y'
cat_features = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'day', 'month', 'poutcome', 'contact']

plt.figure(figsize=(18, 25))
for i, feature in enumerate(cat_features, 1):
    plt.subplot(5, 2, i)
    yes_counts = df[df['y'] == 'yes'][feature].value_counts().sort_values(ascending=False)
    sns.countplot(x=feature, data=df, palette='viridis', order=yes_counts.index)
    plt.title(f'{feature.capitalize()} vs Subscription')
    plt.xticks(rotation=75)
    
plt.suptitle('Distribution of Categorical Features')
plt.tight_layout()

# Save picture as pdf and jpg
plt.savefig('artifact/Distribution_of_Categorical_Features.pdf', format='pdf')
plt.savefig('image/Distribution_of_Categorical_Features.jpg')
plt.show()
```

The phone-call campaigns attracted a larger number of clients to subscribe to term deposits. The managers, technicians and blue-collar workers; married individuals; people without housing and loans; as well as students of secondary and tertiary institutions. The best months of the year for campaigns are April to August and Novermber, while the best days of the week are Thursdays and Fridays.

## Marketing campaign success by analyzing months and days of the week effects

```python
# Subscription rates by month
plt.figure(figsize=(12, 5))

yes_counts = df[df['y'] == 'yes']['month'].value_counts().sort_values(ascending=False)
sns.countplot(x='month', hue='y', data=df, palette='viridis', order=yes_counts.index)
plt.title('Subscription Rates by Month')
plt.xlabel('Month')
plt.ylabel('Count')

# Save picture as pdf and jpg
plt.savefig('artifact/Subscription_Rates_by_Month.pdf', format='pdf')
plt.savefig('image/Subscription_Rates_by_Month.jpg')
plt.show()
```

The best months of the year for campaigns are April to August and Novermber.

```python
# Converting 'day_of_week' to categorical data types
df['day_of_week'] = df['day'].map({1: 'Mon', 2: 'Tue', 3: 'Wed', 4: 'Thu', 5: 'Fri'})

# Analyze subscription rates by day of the week
plt.figure(figsize=(10, 5))

yes_counts = df[df['y'] == 'yes']['day_of_week'].value_counts().sort_values(ascending=False)
sns.countplot(x=df['day_of_week'], hue=df['y'], palette='viridis', order=yes_counts.index) #order=['Mon', 'Tue', 'Wed', 'Thu', 'Fri'])
plt.title('Subscription Rates by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Count')

# Save picture as pdf and jpg
plt.savefig('artifact/Subscription_Rates_by_Day_of_the_Week.pdf', format='pdf')
plt.savefig('image/Subscription_Rates_by_Day_of_the_Week.jpg')
plt.show()
```

The best days of the week are Thursdays and Fridays

## Create customer profiles based on job, marital status, education, and other attributes

```python
# Create a new DataFrame for customers who subscribed (y='yes')
subscribed_df = df[df['y'] == 'yes']

# Define attributes for customer profiling
profile_attributes = ['job', 'marital', 'education', 'housing', 'loan', 'contact', 'poutcome', 'default']

# Visualize customer profiles based on selected attributes
plt.figure(figsize=(11, 20))
for i, attribute in enumerate(profile_attributes, 1):
    plt.subplot(4, 2, i)
    yes_counts = df[df['y'] == 'yes'][attribute].value_counts().sort_values(ascending=False)
    sns.countplot(x=attribute, data=subscribed_df, palette='viridis', order=yes_counts.index)
    plt.title(f'Customer Profile by {attribute}')
    plt.xlabel(attribute)
    plt.ylabel('Count')
    plt.xticks(rotation=80)

plt.tight_layout()

# Save picture as pdf and jpg
plt.savefig('artifact/Customer_Profile_by_Profiling.pdf', format='pdf')
plt.savefig('image/Customer_Profile_by_Profiling.jpg')
plt.show()
```

Customers were more managers, married individuals and secondary school students, mostly without housing and loans, who were mostly contacted through cellular phones

## The influence of contact duration and past campaign outcomes on subscription rates

```python
# Filter data for customers who subscribed (y='yes')
subscribed_df = df[df['y'] == 'yes']

# Influence of contact duration on subscription rates
plt.figure(figsize=(12, 6))

sns.boxplot(x='y', y='duration', data=df, palette='viridis')
plt.title('Influence of Contact Duration on Subscription Rates')
plt.xlabel('Subscription')
plt.ylabel('Contact Duration')

# Save picture as pdf and jpg
plt.savefig('artifact/Influence_of_Contact_Duration_on_Subscription_Rates.pdf', format='pdf')
plt.savefig('image/Influence_of_Contact_Duration_on_Subscription_Rates.jpg')
plt.show()
```

The longer the contact duration with the client, the higher the rate of subscription to term deposits

```python
# Getting the average duration of the last contacts influencing the subscription decision
mean_duration = df[df['contact'] == 'cellular']['duration'].mean()

print("Mean duration of cellular contacts:", mean_duration)
```

```python
# Influence of past campaign outcomes on subscription rates
plt.figure(figsize=(12, 6))

yes_counts = df[df['y'] == 'yes']['poutcome'].value_counts().sort_values(ascending=False)
sns.countplot(x='poutcome', hue='y', data=df, palette='viridis', order=yes_counts.index)
plt.title('Influence of Past Campaign Outcomes on Subscription Rates')
plt.xlabel('Previous Campaign Outcome')
plt.ylabel('Count')

# Save picture as pdf and jpg
plt.savefig('artifact/Influence_of_Past_Campaign_Outcomes_on_Subscription_Rates.pdf', format='pdf')
plt.savefig('image/Influence_of_Past_Campaign_Outcomes_on_Subscription_Rates.jpg')
plt.show()
```

Only successful campaigns had more subscriptions to term deposit

```python
# Getting the average duration of the last contacts influencing the subscription decision
mean_duration = df[df['contact'] == 'cellular']['duration'].mean()

print("Mean duration of cellular contacts:", mean_duration)
```

# Conclusion

This project identified key campaign parameters and relevant factors that significantly influence client subscriptions to term deposit, which will help the banking institutions tailor their marketing efforts effectively. The findings revealed;

- That phone-call campaigns attracted a larger number of clients to subscribe to term deposits.
- That managers, technicians and blue-collar workers; married individuals; people without housing and loans; as well as students of secondary and tertiary institutions.
- That the best months of the year for campaigns are April to August, while the best days of the week are Thursdays and Fridays.
- That the longer the contact duration with the client, the higher the rate of subscription to term deposits.

# Recommendations
From the findings of this project, financial institutions will have more clients subscribing to term deposit if they:
- enploy more of phone-call campaigns for an increased subscription of clients to term deposit.
- channel more of these phone-call campaigns to managers, technicians and blue-collar workers; married and single individuals; people with and without housing and loans; as well as students of secondary and tertiary institutions.
- intensify their phone-call campaigns on Thursdays and Fridays, and from April to August.
- keep their phone-call campaigns contact at an average of 263 seconds per contact.
